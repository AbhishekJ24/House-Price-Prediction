```python
import pandas as pd
import numpy as np
import random
import math
```

```python
df = pd.read_csv('before.csv')
df.head()
```

##### The above dataset has alot of problems which need to be engineered as per our requirements

```python
floors = df['number of floors']
floors
```

```python
new_floors=[]
for floor in floors:
    new_floors.append(math.ceil(floor))
df['number of floors']=new_floors
df
```

##### Now that the floors are fixed, we can move towards fixing bathrooms values

```python
baths = df['bathrooms']
baths
```

```python
new_bathrooms=[]
for bath in baths:
    new_bathrooms.append(math.ceil(bath))
df['bathrooms']=new_bathrooms
df
```
